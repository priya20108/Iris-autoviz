```python
# A vertical boxplot of sepal lenght  grouped by various species:
sns.set_style("whitegrid")
  
sns.boxplot(x ='Species', y ='SepalLengthCm', data =df)
```

take the first box plot i.e, blue box plot of the figure and understand these statistical things:
*Bottom black horizontal line of blue box plot is minimum value
*First black horizontal line of rectangle shape of blue box plot is First 
 quartile or 25%
*Second black horizontal line of rectangle shape of blue box plot is Second     quartile or 50% or median.
*Third black horizontal line of rectangle shape of blue box plot is third quartile or 75%
*Top black horizontal line of rectangle shape of blue box plot is maximum value.
*Small diamond shape of blue box plot is outlier data or erroneous data
*By obsarving it we can say setosa has less sepal lenght then versicolor and verginica

```python
#violinplot for visualizing petal length for various species
sns.violinplot(x="Species",y="PetalLengthCm", data=df, size=8)
plt.show()
```

A violin plot is a method of plotting numeric data. It is similar to a box plot, with the addition of a rotated kernel density plot on each side. Violin plots are similar to box plots, except that they also show the probability density of the data at different values, usually smoothed by a kernel density estimator.
